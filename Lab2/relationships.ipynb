```python
import numpy as np
import pandas as pd
import seaborn as sns
import glob
from matplotlib import pyplot as plt
```

# DT

# Блоками

```python
data_ms = {
    "matrix_size": [16, 64, 256, 512],
    "elapsed_time": [0.013511, 0.035121, 0.066322, 0.125357],
}  # proccesses_num = 4

data_pn = {
    "proccesses_num": [1, 2, 4, 8],
    "elapsed_time": [0.066516, 0.075327, 0.106311, 0.205017],
}   # matrix_size = 512

```

```python
ax = sns.lineplot(data=pd.DataFrame(data_ms), x='matrix_size', y='elapsed_time')
ax.set(xlabel='Matrix Size', ylabel='time')
plt.show()
```

```python
ax = sns.lineplot(data=pd.DataFrame(data_pn), x='proccesses_num', y='elapsed_time')
ax.set(xlabel='Processes num', ylabel='time')
plt.show()
```

## Вывод
1. Время работы не линейно зависет от размера матрицы. Время работы чуть быстрее растет чуть быстрее линейного возврастания
2. Время работы сильно зависет от число процессов.
